## Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

## Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(f"Promedio: {promedio} y Promedio Ponderado: {promedio_ponderado}")
```

## Mediana

```python
mediana = np.median(data)
print(media)
```

## Varianza y Desviaci칩n Est치ndar

```python
varianza = np.var(data)
desviacion = np.std(data)
print(f"Varianza: {varianza}, Desviacion: {desviacion}")
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t
```

```python
tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```

```python
from enum import Enum

class Color(Enum):
    accent  = '#2563eb'  # azul
    accent2 = '#16a34a'  # verde
    accent3 = '#f59e0b'  # 치mbar
    neutral = '#6b7280'  # gris
```

```python
fig, ax = plt.subplots(figsize=(8,5))
bins = range(int(df['edad'].min())-1, int(df['edad'].max())+2)
ax.hist(df['edad'], bins=bins, color=Color.accent.value, edgecolor='white', alpha=0.85)
mean_age = df['edad'].mean()
med_age = df['edad'].median()
ax.axvline(mean_age, color=Color.accent2.value, linestyle='--', linewidth=2, label=f"Media: {mean_age:.2f}")
ax.axvline(med_age, color=Color.accent3.value, linestyle=':', linewidth=2.5, label=f"Mediana: {med_age:.2f}")
ax.set_title('Distribuci칩n de edades')
ax.set_xlabel('Edad (a침os)')
ax.set_ylabel('Frecuencia')
ax.legend()
plt.tight_layout()
plt.show()
fig.savefig('img\\01_hist_edades.png', dpi=180)
```

```python
# =======================
# 2) Boxplot de edades
# =======================
fig, ax = plt.subplots(figsize=(6,4))
ax.boxplot(df['edad'], vert=False, patch_artist=True,
           boxprops=dict(facecolor=Color.accent.value, color=Color.accent.value),
           medianprops=dict(color='white', linewidth=2),
           whiskerprops=dict(color=Color.accent.value), capprops=dict(color=Color.accent.value))
ax.set_title('Caja y bigotes: edad')
ax.set_xlabel('Edad (a침os)')
plt.tight_layout()
plt.show()
fig.savefig('img\\02_box_edades.png', dpi=180)
```

```python
# ====================================================
# 3) Barras: frecuencias de S칤/No para Python/Colab/GitHub
# ====================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Program칩 en Python','Us칩 Google Colab','Cuenta de GitHub'],
    [Color.accent.value, Color.accent2.value, Color.accent3.value]
):
    counts = df[col].value_counts().reindex([True, False]).fillna(0).astype(int)
    labels = ['S칤','No']
    ax.bar(labels, counts, color=[color, Color.neutral.value], edgecolor='white')
    ax.set_title(title)
    ax.set_ylim(0, max(counts.max(), 1)+2)
    for i, v in enumerate(counts):
        ax.text(i, v+0.1, str(v), ha='center', va='bottom', fontsize=9)
axes[0].set_ylabel('N칰mero de estudiantes')
plt.tight_layout()
plt.show()
fig.savefig('img\\03_barras_frecuencias.png', dpi=180)
```

```python
# =======================================
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2칑2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','S칤'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','S칤'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'img\\04_heatmap_{idx}.png', dpi=180)
```

```python
# ======================================================
# 5) Barras: edad promedio por grupo (S칤/No) por variable
# ======================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Edad promedio por Python','Edad promedio por Colab','Edad promedio por GitHub'],
    [Color.accent.value, Color.accent2.value, Color.accent3.value]
):
    means = df.groupby(col)['edad'].mean().reindex([True, False])
    labels = ['S칤','No']
    bars = ax.bar(labels, means, color=[color, Color.neutral.value], edgecolor='white')
    for i, v in enumerate(means):
        ax.text(i, v+0.05, f"{v:.2f}", ha='center', va='bottom', fontsize=9)
    ax.set_title(title)
    ax.set_xlabel('Grupo')
axes[0].set_ylabel('Edad promedio (a침os)')
plt.tight_layout()
plt.show()
fig.savefig('img\\05_barras_edad_promedio.png', dpi=180)
```
