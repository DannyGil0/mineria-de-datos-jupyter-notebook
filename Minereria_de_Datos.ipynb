````python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t
````

# MSE: Mean Squared Error y otros indicadores de desempeño de modelos

````python
import numpy as np
import pandas as pd

anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
````

Primer intento de modelar los años de escolaridad y el ingreso estimado.
Utilizar el primer punto y el útlimo para trazar una recta:
$$ m = \frac{y_{2} - y_{1}}{x_{2} - x_{1}} $$
$$ b = y_{1} - m * x_{1} $$
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

````python

x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
````

````python
import matplotlib.pyplot as plt
# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
````

Valores predichos $(y_{pred})$: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.


Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error

$$ MSE = \frac {\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n} $$
Donde:
$ MSE = $ error cuadrático medio.
$ n = $ número de datos.
$ Y_{i} = $ valores observados.
$ \hat{Y_{i}} = $ valores predichos.

````python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
````

````python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
````

````python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
````

````python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
````

````python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
````

Ahora comparemos el MSE de una regresión lineal.
Por ahora no es importante el detalle de la regrsión lineal.
Lo importante es comparar el MSE.

````python
from sklearn.linear_model import LinearRegression
import numpy as np

# Cambia la forma de anios_escolaridad a 2D array como es requerido por sklearn
X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx

X
# Crear y ajustar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener los coeficientes (m) e intercepto (b) del nuevo modelo
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

print(f"Coefficients (m) from scikit-learn: {m_sklearn:.2f}")
print(f"Intercept (b) from scikit-learn: {b_sklearn:.2f}")
````

Además del MSE (Error Cuadrático Medio), existen varios indicadores clave para evaluar modelos de regresión lineal, cada uno con un enfoque distinto sobre el rendimiento del modelo:



*   **RMSE** (Raíz del Error Cuadrático Medio): Es la raíz cuadrada del MSE. A diferencia del MSE, el RMSE expresa el error en las mismas unidades que la variable objetivo, lo que facilita su interpretación. Es útil para penalizar errores grandes, similar al MSE.
*   **MAE** (Error Absoluto Medio): Calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales. Es más robusto ante valores atípicos (outliers) que el MSE, ya que no eleva al cuadrado los errores grandes.
*   **$R^{2}$** (Coeficiente de Determinación): Indica qué proporción de la variación de la variable dependiente es explicada por el modelo. Va de 0 a 1; un valor más cercano a 1 indica un mejor ajuste.
* **$R^{2}$** Ajustado: Similar al **$R^{2}$**, pero ajusta el valor según el número de predictores (variables independientes) en el modelo. Es más preciso que el **$R^{2}$** cuando se añaden variables innecesarias, ya que penaliza la complejidad excesiva.
*   **MAPE** (Error Porcentual Absoluto Medio): Mide el error en términos porcentuales, lo que facilita entender la precisión relativa del modelo (ej. "el modelo se equivoca en un 5% en promedio").

### Fórmulas de los Indicadores de Regresión

**RMSE (Root Mean Square Error) - Raíz del Error Cuadrático Medio:**
$$ RMSE = \sqrt{\frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n}} = \sqrt{MSE} $$

**MAE (Mean Absolute Error) - Error Absoluto Medio:**
$$ MAE = \frac{\Sigma_{i=1}^{n}|Y_{i}-\hat{Y_{i}}|}{n} $$

**$R^{2}$ (Coefficient of Determination) - Coeficiente de Determinación:**
$$ R^{2} = 1 - \frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{\Sigma_{i=1}^{n}(Y_{i}-\bar{Y})^{2}} $$

Donde $\bar{Y}$ es la media de los valores observados.

**$R^{2}$ Ajustado (Adjusted R-squared):**
$$ R^{2}_{ajustado} = 1 - \frac{(1-R^{2})(n-1)}{n-p-1} $$

Donde $p$ es el número de variables independientes (predictores).

**MAPE (Mean Absolute Percentage Error) - Error Porcentual Absoluto Medio:**
$$ MAPE = \frac{100\%}{n} \times \Sigma_{i=1}^{n}\left|\frac{Y_{i}-\hat{Y_{i}}}{Y_{i}}\right| $$

````python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Metricas
mae = mean_absolute_error(ingreso_mensual_estimado_mx, y_pred_sklearn)
mse = mean_squared_error(ingreso_mensual_estimado_mx, y_pred_sklearn)
rmse = np.sqrt(mse)
r2 = r2_score(ingreso_mensual_estimado_mx, y_pred_sklearn)
mape = np.mean(np.abs((ingreso_mensual_estimado_mx - y_pred_sklearn) / ingreso_mensual_estimado_mx)) * 100

# R2 ajustado (1 predictor)
n = len(ingreso_mensual_estimado_mx)
p = 1
r2_ajustado = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Resultados
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R^2:", r2)
print("R^2 Ajustado:", r2_ajustado)
print("MAPE:", mape)
````

Resumen de uso:
* Precisión general y sensibilidad a outliers: MAE.
* Penalizar errores grandes: RMSE o MSE.
* Explicación del modelo: $R^{2}$ o $R^{2}$ Ajustado.
* Interpretación porcentual: MAPE.

````python
def resumen_razon(s):
    return pd.Series({
    "n": s.count(),
    "media": s.mean(),
    "mediana": s.median(),
    "sd": s.std(),
    "min": s.min(),
    "q1": s.quantile(0.25),
    "q3": s.quantile(0.75),
    "max": s.max(),
    "cv": s.std()/s.mean() if s.mean()!=0 else np.nan
    }).round(2)

razon_stats = pd.concat({
    "edad": resumen_razon(df["edad"]),
    "ingreso_mensual": resumen_razon(df["ingreso_mensual"]),
    "visitas_ultimo_anio": resumen_razon(df["visitas_ultimo_anio"]),
    "dias_desde_ultima_visita": resumen_razon(df["dias_desde_ultima_visita"])
    }, axis=1)
razon_stats
````

````python
features = pd.read_csv('src/mondrian-art-project/mondrian-painting-features.csv')
painting_info = pd.read_csv('src/mondrian-art-project/mondrian-painting-info.csv')

# Compute width and height per painting
dims = features.groupby('painting_id').apply(
    lambda df: pd.Series({
        'paint_width':  (df['x'] + df['width']).max(),
        'paint_height': (df['y'] + df['height']).max()
    })
).reset_index()
````

````python
dims['paint_area'] = dims['paint_width'] * dims['paint_height']
painting_info = painting_info.merge(dims, on='painting_id', how='left')
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, col, label in zip(
    axes,
    ['paint_width', 'paint_height', 'paint_area'],
    ['Width (px)', 'Height (px)', 'Area (px²)']
):
    ax.scatter(painting_info['year'], painting_info[col], alpha=0.7)
    ax.set_xlabel('Year')
    ax.set_ylabel(label)
    ax.set_title(f'Painting {label} Over Time')

plt.tight_layout()
plt.show()
````

Extensión 2 – Uso de colores a lo largo del tiempo:

````python
class PlotColor(Enum):
    blue   = '#2f5a9f'
    red    = '#e6151a'
    yellow = '#d4c600'
    white  = '#aaaaaa'
    black  = '#222222'
    gray   = '#888888'
````

````python
# Asegura que feature_area y total_area existan
features['feature_area'] = features['width'] * features['height']
painting_info['total_area'] = painting_info['paint_width'] * painting_info['paint_height']

colors = [c.name for c in PlotColor]
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for ax, color in zip(axes.flat, colors):
    color_feats = features[features['color'] == color]

    color_area = (
        color_feats.groupby('painting_id')['feature_area']
        .sum()
        .reset_index(name=f'{color}_area')
    )

    merged = painting_info.merge(color_area, on='painting_id', how='left')
    merged[f'{color}_area'] = merged[f'{color}_area'].fillna(0)
    merged[f'{color}_pct'] = 100 * merged[f'{color}_area'] / merged['total_area']

    ax.scatter(
        merged['year'], merged[f'{color}_pct'],
        color=PlotColor[color].value, alpha=0.75, edgecolors='black', linewidths=0.3
    )
    ax.set_xlabel('Year')
    ax.set_ylabel('% of painting area')
    ax.set_title(f'{color.capitalize()} usage over time')

plt.suptitle('Color Usage in Mondrian Paintings (1920–1940)', fontsize=14)
plt.tight_layout()
plt.show()
````

````python
print("\nMerged data for correlation analysis:")
print(all_languages)
````

````python
# Create scatter plots to analyze correlations between languages
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# English vs Spanish
axes[0].scatter(all_languages['percentage_english'], all_languages['percentage_spanish'], alpha=0.7)
axes[0].set_xlabel('English Percentage (%)')
axes[0].set_ylabel('Spanish Percentage (%)')
axes[0].set_title('English vs Spanish Color Naming')

# Add color labels for significant points
for i, row in all_languages.iterrows():
    if row['percentage_english'] > 5 or row['percentage_spanish'] > 5:  # Only label significant colors
        axes[0].annotate(row['color_name'],
                        (row['percentage_english'], row['percentage_spanish']),
                        xytext=(5, 5), textcoords='offset points', fontsize=8)

# English vs Tsimane
axes[1].scatter(all_languages['percentage_english'], all_languages['percentage_tsimane'], alpha=0.7)
axes[1].set_xlabel('English Percentage (%)')
axes[1].set_ylabel('Tsimane Percentage (%)')
axes[1].set_title('English vs Tsimane Color Naming')

for i, row in all_languages.iterrows():
    if row['percentage_english'] > 5 or row['percentage_tsimane'] > 5:
        axes[1].annotate(row['color_name'],
                        (row['percentage_english'], row['percentage_tsimane']),
                        xytext=(5, 5), textcoords='offset points', fontsize=8)

# Spanish vs Tsimane
axes[2].scatter(all_languages['percentage_spanish'], all_languages['percentage_tsimane'], alpha=0.7)
axes[2].set_xlabel('Spanish Percentage (%)')
axes[2].set_ylabel('Tsimane Percentage (%)')
axes[2].set_title('Spanish vs Tsimane Color Naming')

for i, row in all_languages.iterrows():
    if row['percentage_spanish'] > 5 or row['percentage_tsimane'] > 5:
        axes[2].annotate(row['color_name'],
                        (row['percentage_spanish'], row['percentage_tsimane']),
                        xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

# Calculate correlation coefficients
correlation_en_es = all_languages['percentage_english'].corr(all_languages['percentage_spanish'])
correlation_en_ts = all_languages['percentage_english'].corr(all_languages['percentage_tsimane'])
correlation_es_ts = all_languages['percentage_spanish'].corr(all_languages['percentage_tsimane'])
````
