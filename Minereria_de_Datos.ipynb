````python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t
````

# MSE: Mean Squared Error y otros indicadores de desempe침o de modelos

````python
import numpy as np
import pandas as pd

anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('A침os de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('A침os de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
````

Primer intento de modelar los a침os de escolaridad y el ingreso estimado.
Utilizar el primer punto y el 칰tlimo para trazar una recta:
$$ m = \frac{y_{2} - y_{1}}{x_{2} - x_{1}} $$
$$ b = y_{1} - m * x_{1} $$
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

````python

x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuaci칩n de la recta es: y = {m:.2f}x + {b:.2f}")
````

````python
import matplotlib.pyplot as plt
# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresi칩n (y = {m:.2f}x + {b:.2f})')
plt.title('A침os de Escolaridad vs. Ingreso Mensual Estimado con L칤nea de Regresi칩n')
plt.xlabel('A침os de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
````

Valores predichos $(y_{pred})$: Son los ingresos mensuales que nuestra recta de regresi칩n predice para cada a침o de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicci칩n fue menor que el valor real, y un error negativo indica que la predicci칩n fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre s칤) y para penalizar m치s los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la l칤nea.

Error Cuadr치tico Medio (MSE): Es la suma de los errores cuadrados dividida por el n칰mero total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE m치s bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadr치tico Medio (MSE) es de aproximadamente 13,196,667.31. Este n칰mero nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la l칤nea.


F칩rmula del MSE (Error Cuadr치tico Medio) del Ingl칠s Mean Squared Error

$$ MSE = \frac {\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n} $$
Donde:
$ MSE = $ error cuadr치tico medio.
$ n = $ n칰mero de datos.
$ Y_{i} = $ valores observados.
$ \hat{Y_{i}} = $ valores predichos.

````python
# 1. Calcular los valores predichos (y_pred) usando la ecuaci칩n de la recta
y_pred = m * anios_escolaridad + b
y_pred
````

````python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
````

````python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
````

````python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
````

````python
# 5. Dividir la suma de los errores cuadrados por el n칰mero de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
````

Ahora comparemos el MSE de una regresi칩n lineal.
Por ahora no es importante el detalle de la regrsi칩n lineal.
Lo importante es comparar el MSE.

````python
from sklearn.linear_model import LinearRegression
import numpy as np

# Cambia la forma de anios_escolaridad a 2D array como es requerido por sklearn
X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx

X
# Crear y ajustar el modelo de regresi칩n lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener los coeficientes (m) e intercepto (b) del nuevo modelo
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

print(f"Coefficients (m) from scikit-learn: {m_sklearn:.2f}")
print(f"Intercept (b) from scikit-learn: {b_sklearn:.2f}")
````

El nuevo modelo de recta obtenido con regrsi칩n lineal es:
$$ y = 1914.55 * x + 9731.66 $$

````python
import matplotlib.pyplot as plt

# valores predichos por sklearn
y_pred_sklearn = m_sklearn * anios_escolaridad + b_sklearn

# grafica
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, color='blue', label='Datos')
plt.plot(anios_escolaridad, m * anios_escolaridad + b, color='red', label='Regresi칩n Simple')
plt.plot(anios_escolaridad, y_pred_sklearn, color='green', linestyle='--', label='Regresi칩n sklearn')
plt.legend()
plt.show()
````

````python
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn
mse_sklearn = np.mean(errores_sklearn**2)

print("MSE Simple:", mse)
print("MSE sklearn:", mse_sklearn)
print("Diferencia:", abs(mse - mse_sklearn))
print("Mejora %:", ((mse - mse_sklearn) / mse) * 100)
if mse_sklearn < mse:
    print("sklearn menor error")
else:
    print("simple menor error")

print("Diferencia:", abs(mse - mse_sklearn))
print("Mejora %:", ((mse - mse_sklearn) / mse) * 100)
````

Adem치s del MSE (Error Cuadr치tico Medio), existen varios indicadores clave para evaluar modelos de regresi칩n lineal, cada uno con un enfoque distinto sobre el rendimiento del modelo:



*   **RMSE** (Ra칤z del Error Cuadr치tico Medio): Es la ra칤z cuadrada del MSE. A diferencia del MSE, el RMSE expresa el error en las mismas unidades que la variable objetivo, lo que facilita su interpretaci칩n. Es 칰til para penalizar errores grandes, similar al MSE.
*   **MAE** (Error Absoluto Medio): Calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales. Es m치s robusto ante valores at칤picos (outliers) que el MSE, ya que no eleva al cuadrado los errores grandes.
*   **$R^{2}$** (Coeficiente de Determinaci칩n): Indica qu칠 proporci칩n de la variaci칩n de la variable dependiente es explicada por el modelo. Va de 0 a 1; un valor m치s cercano a 1 indica un mejor ajuste.
* **$R^{2}$** Ajustado: Similar al **$R^{2}$**, pero ajusta el valor seg칰n el n칰mero de predictores (variables independientes) en el modelo. Es m치s preciso que el **$R^{2}$** cuando se a침aden variables innecesarias, ya que penaliza la complejidad excesiva.
*   **MAPE** (Error Porcentual Absoluto Medio): Mide el error en t칠rminos porcentuales, lo que facilita entender la precisi칩n relativa del modelo (ej. "el modelo se equivoca en un 5% en promedio").

### F칩rmulas de los Indicadores de Regresi칩n

**RMSE (Root Mean Square Error) - Ra칤z del Error Cuadr치tico Medio:**
$$ RMSE = \sqrt{\frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n}} = \sqrt{MSE} $$

**MAE (Mean Absolute Error) - Error Absoluto Medio:**
$$ MAE = \frac{\Sigma_{i=1}^{n}|Y_{i}-\hat{Y_{i}}|}{n} $$

**$R^{2}$ (Coefficient of Determination) - Coeficiente de Determinaci칩n:**
$$ R^{2} = 1 - \frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{\Sigma_{i=1}^{n}(Y_{i}-\bar{Y})^{2}} $$

Donde $\bar{Y}$ es la media de los valores observados.

**$R^{2}$ Ajustado (Adjusted R-squared):**
$$ R^{2}_{ajustado} = 1 - \frac{(1-R^{2})(n-1)}{n-p-1} $$

Donde $p$ es el n칰mero de variables independientes (predictores).

**MAPE (Mean Absolute Percentage Error) - Error Porcentual Absoluto Medio:**
$$ MAPE = \frac{100\%}{n} \times \Sigma_{i=1}^{n}\left|\frac{Y_{i}-\hat{Y_{i}}}{Y_{i}}\right| $$

````python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Metricas
mae = mean_absolute_error(ingreso_mensual_estimado_mx, y_pred_sklearn)
mse = mean_squared_error(ingreso_mensual_estimado_mx, y_pred_sklearn)
rmse = np.sqrt(mse)
r2 = r2_score(ingreso_mensual_estimado_mx, y_pred_sklearn)
mape = np.mean(np.abs((ingreso_mensual_estimado_mx - y_pred_sklearn) / ingreso_mensual_estimado_mx)) * 100

# R2 ajustado (1 predictor)
n = len(ingreso_mensual_estimado_mx)
p = 1
r2_ajustado = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# Resultados
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R^2:", r2)
print("R^2 Ajustado:", r2_ajustado)
print("MAPE:", mape)
````

Resumen de uso:
* Precisi칩n general y sensibilidad a outliers: MAE.
* Penalizar errores grandes: RMSE o MSE.
* Explicaci칩n del modelo: $R^{2}$ o $R^{2}$ Ajustado.
* Interpretaci칩n porcentual: MAPE.

````python
np.random.seed(42)
n = 200
````

### Crea el DataFrame (o carga desde CSV si lo prefieres)

````python
df = pd.DataFrame({
    "id_cliente": range(1, n+1),
    # Identificador unico para cada cliente (1 a n)

    "genero": np.random.choice(
        ["F", "M", "No binario", "Prefiero no decir"],
        size=n, p=[0.45, 0.45, 0.05, 0.05]
    ),
    # Genero del cliente, con probabilidades asignadas

    "segmento": np.random.choice(
        ["Est치ndar", "Oro", "Platino"],
        size=n, p=[0.6, 0.3, 0.1]
    ),
    # Segmento de membresia: mayoria estandar, menos oro y platino

    "satisfaccion": np.random.choice(
        [1,2,3,4,5],
        size=n, p=[0.1,0.2,0.35,0.25,0.1]
    ),
    # Nivel de satisfaccion (1-5), mas frecuente en valores intermedios

    "ciudad": np.random.choice(
        ["SLP", "CDMX", "GDL", "MTY", "QRO"],
        size=n, p=[0.35,0.25,0.15,0.15,0.10]
    ),
    # Ciudad de residencia, con mayor peso en SLP y CDMX

    "edad": np.random.normal(35, 10, size=n).round(0).clip(18, 70).astype(int),
    # Edad simulada con distribucion normal (media 35), limitada entre 18 y 70

    "ingreso_mensual": np.random.lognormal(mean=9.5, sigma=0.5, size=n).round(2),
    # Ingreso mensual en pesos MXN, distribuci칩n log-normal para valores positivos

    "temperatura_sala": np.random.normal(23, 2, size=n).round(1),
    # Temperatura promedio en sala, normal con media 23춿C

    "visitas_ultimo_anio": np.random.poisson(lam=3, size=n),
    # N칰mero de visitas en el 칰ltimo a침o, distribuci칩n Poisson con media 3

    "fecha_ultima_visita": pd.to_datetime("2025-12-31")
        - pd.to_timedelta(np.random.randint(0, 365, size=n), unit="D")
    # Fecha de 칰ltima visita, dentro del a침o 2025
})
df.head()
````

````python
df.dtypes
````

### Tipado correcto por escala
- Nominal: genero, ciudad, id_cliente (como identificador, no lo analices).
- Ordinal: segmento, satisfaccion (con orden definido).
- Intervalo: temperatura_sala (춿C tiene 0 arbitrario; diferencias s칤 tienen sentido).
- Raz칩n: edad, ingreso_mensual, visitas_ultimo_anio (tienen cero absoluto y proporciones).

````python
# Nominal  category (sin orden)
df["genero"] = df["genero"].astype("category")
df["ciudad"] = df["ciudad"].astype("category")

# Ordinal  category con orden
segmento_cat = pd.CategoricalDtype(categories=["Est치ndar", "Oro", "Platino"], ordered=True)
df["segmento"] = df["segmento"].astype(segmento_cat)

satisfaccion_cat = pd.CategoricalDtype(categories=[1,2,3,4,5], ordered=True)
df["satisfaccion"] = df["satisfaccion"].astype(satisfaccion_cat)

# Intervalo / Raz칩n  num칠rico (ya est치n), pero verifica:
df["temperatura_sala"] = pd.to_numeric(df["temperatura_sala"])
df["edad"] = pd.to_numeric(df["edad"])
df["ingreso_mensual"] = pd.to_numeric(df["ingreso_mensual"])
df["visitas_ultimo_anio"] = pd.to_numeric(df["visitas_ultimo_anio"])

# Derivada 칰til: d칤as desde 칰ltima visita (raz칩n)
df["dias_desde_ultima_visita"] = (pd.to_datetime("2025-12-31") - df["fecha_ultima_visita"]).dt.days
````

Raz칩n (media, mediana, IQR, coeficiente de variaci칩n)

````python
def resumen_razon(s):
    return pd.Series({
    "n": s.count(),
    "media": s.mean(),
    "mediana": s.median(),
    "sd": s.std(),
    "min": s.min(),
    "q1": s.quantile(0.25),
    "q3": s.quantile(0.75),
    "max": s.max(),
    "cv": s.std()/s.mean() if s.mean()!=0 else np.nan
    }).round(2)

razon_stats = pd.concat({
    "edad": resumen_razon(df["edad"]),
    "ingreso_mensual": resumen_razon(df["ingreso_mensual"]),
    "visitas_ultimo_anio": resumen_razon(df["visitas_ultimo_anio"]),
    "dias_desde_ultima_visita": resumen_razon(df["dias_desde_ultima_visita"])
    }, axis=1)
razon_stats
````
