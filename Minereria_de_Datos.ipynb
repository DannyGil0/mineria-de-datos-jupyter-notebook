## Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

## Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(f"Promedio: {promedio} y Promedio Ponderado: {promedio_ponderado}")
```

## Mediana

```python
mediana = np.median(data)
print(media)
```

## Varianza y Desviaci칩n Est치ndar

```python
varianza = np.var(data)
desviacion = np.std(data)
print(f"Varianza: {varianza}, Desviacion: {desviacion}")
```

```python
# Frecuencias (NumPy)
freq_python = {'Si': int(np.count_nonzero(prog_py)), 'No': int(edad.size - np.count_nonzero(prog_py))}
freq_colab  = {'Si': int(np.count_nonzero(usa_colab)), 'No': int(edad.size - np.count_nonzero(usa_colab))}
freq_gh     = {'Si': int(np.count_nonzero(cuenta_gh)), 'No': int(edad.size - np.count_nonzero(cuenta_gh))}
```

```python
print(f"Frecuencias de utilizar Python: {freq_python}")
print(f"Frecuencias de utilizar Colab: {freq_colab}")
print(f"Frecuencias de utilizar GitHub: {freq_gh}")
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t
```

```python
tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```

```python
print(f"Tabla cruzada entre Python y Colab: {tab_py_colab}")
print(f"Tabla cruzada entre Python y GitHub: {tab_py_gh}")
print(f"Tabla cruzada entre Colab y GitHub: {tab_colab_gh}")
```
