```python
import pandas as pd
df = pd.read_csv("src/Examen Parcial 2/items.csv")

df.eval("importe = precio * cantidad", inplace=True)

dfa = df.query("categoria == 'A'")

promedio_por_categoria = df.groupby("categoria")["importe"].mean()

# Respuesta: promedio de categoría A (redondeado a 3 decimales)
respuesta = round(promedio_por_categoria["A"], 3)
print(f"Promedio de importe para categoría A: {respuesta}")
```

```python
import pandas as pd

# 1. Lee ambos CSVs en dfc (clientes) y dfo (órdenes)
dfc = pd.read_csv("src/Examen Parcial 2/clientes.csv")  # clientes
dfo = pd.read_csv("src/Examen Parcial 2/ordenes.csv")   # órdenes

resultado = dfc.merge(dfo, on="cliente_id", how="left")

# (a) Número total de filas del resultado del left merge
total_filas = len(resultado)
print(f"(a) Total de filas: {total_filas}")

# (b) Monto total por cliente (suma de monto) y reemplaza NaN por 0
monto_por_cliente = resultado.groupby("cliente_id")["monto"].sum().fillna(0)
print(f"\n(b) Monto total por cliente:\n{monto_por_cliente}")

# Respuesta: monto total para cliente_id=1
respuesta = monto_por_cliente.get(1, 0)
print(f"\nMonto total para cliente_id=1: {respuesta}")
```

```python
import pandas as pd
df = pd.read_csv("src/Examen Parcial 2/sensores.csv")
df = df[['ts', 'temperatura', 'humedad']]

df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')

# (a) ¿Cuántos NaT quedaron en ts?
a = df['ts'].isna().sum()

df['temperatura'] = df['temperatura'].fillna(df['temperatura'].median())
df['humedad'] = df['humedad'].fillna(df['humedad'].mean())

# (b) Valor final de la mediana de temperatura tras imputación (redondea a 1 decimal)
b = round(df['temperatura'].median(), 1)

# (c) Valor final de la media de humedad tras imputación (redondea a 3 decimales)
c = round(df['humedad'].mean(), 3)

print(f"Respuestas: {a}, {b}, {c}")
```
